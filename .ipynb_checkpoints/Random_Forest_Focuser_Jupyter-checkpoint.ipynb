```python
def remove_nulls_and_non_numeric(target_df, feature_df):
    try:
        # Combine the dataframes
        combined_df = pd.concat([target_df, feature_df], axis=1) 
        original_length = len(combined_df)

        # Put the column names into variables 
        target_column_name = combined_df.columns[0]
        feature_column_name = combined_df.columns[1]

        # Initialize counters for non-numeric and null values in the columns
        non_numeric_count = {target_column_name: 0, feature_column_name: 0}
        null_count = {target_column_name: 0, feature_column_name: 0} 

        # Iterate through each column to count non-numeric and null values
        for column in combined_df.columns:
            # Convert to numeric, coercing non-numeric to NaN
            numeric_column = pd.to_numeric(combined_df[column], errors='coerce')

            # Counting non-numeric values (original non-NaN non-numeric values will become NaN)
            non_numeric_count[column] = numeric_column.isna().sum() - combined_df[column].isna().sum()

            # Counting null values (original NaN values)
            null_count[column] = combined_df[column].isnull().sum()

        # Print the counts of non-numeric and null values
        #print("Non-numeric value count per column:")
        #for col, count in non_numeric_count.items():
        #    print(f"{col}: {count}")

        #print("\nNull value count per column:")
        #for col, count in null_count.items():
        #    print(f"{col}: {count}")

        # Convert non-numeric values to NaN for each column
        for column in combined_df.columns:
            combined_df[column] = pd.to_numeric(combined_df[column], errors='coerce')

        # Drop rows with NaN values in either column
        combined_df.dropna(inplace=True) 

        # Get new length of dataframe 
        processed_length = len(combined_df) 
        print(f"The processed dataframe removed {original_length - processed_length} values with non-numeric or null values. The processed dataframe has {processed_length} data pairs.")

        # Return results 
        processed_target_df = combined_df[[target_column_name]].copy()
        processed_feature_df = combined_df[[feature_column_name]].copy() 
        return processed_target_df, processed_feature_df
        
    except Exception as e:
        print(f"There was an error when executing the remove_nulls_and_non_numeric() function. Error: {e}")

```

```python
# Autobucket t-test  
def autobucket_t_test(feature_df, target_df):

    # Get the average, median and blended average of target
    target_mean = target_df.iloc[:, 0].mean()
    target_median = target_df.iloc[:, 0].median()
    blended_average = ((target_mean + target_median) / 2)
    

    # Put feature variables into high vs low. 
    features_with_low_target = []
    features_with_high_target = []
    
    i = 0
    while i < len(feature_df):
        # Select the feature and target value to analyze
        feature_value = feature_df.iloc[i, 0]
        target_value = target_df.iloc[i, 0]

        if target_value >= blended_average:
            features_with_high_target.append(feature_value) 
        else:
            features_with_low_target.append(feature_value)
        i += 1

    
    number_of_high_target_features = len(features_with_high_target)
    number_of_low_target_features = len(features_with_low_target)
    avg_feature_value_high_target = sum(features_with_high_target)/len(features_with_high_target)
    avg_feature_value_low_target = sum(features_with_low_target)/len(features_with_low_target)

    # Perform Welch's t-test
    t_statistic, p_value = ttest_ind(features_with_high_target, features_with_low_target, equal_var=False)

    results_data = {
    'Threshold': [blended_average],
    'High_Target_Feature_Count': [number_of_high_target_features],
    'Low_Target_Feature_Count': [number_of_low_target_features],
    'High_Target_Feature_Avg': [avg_feature_value_high_target],
    'Low_Target_Feature_Avg': [avg_feature_value_low_target], 
    'T_Statistic': [t_statistic], 
    'p_Value': [p_value]
    }

    return results_data

#autobucket_t_test(feature_df, target_df)
```

```python
# Inital loop to show t test results 
target_df = df_to_use[[target_name]]

# Set max p value and create list of features that have min p value 
max_p_value = 0.01 
features_passed_t_test = [] 
features_failed_t_test = []
feature_t_test_results = pd.DataFrame(columns=['feature', 'pass_fail', 'p_value', 't_statistic', 'high_target_avg', 'low_target_avg'])

for feature in feature_list:
    print(f"Processing feature: {feature}")
    
    # Create single column feature_df
    feature_df = df_to_use[[feature]]
    
    # Remove non-numeric and null values 
    cleanup_results = remove_nulls_and_non_numeric(target_df, feature_df)
    processed_target_df = cleanup_results[0] 
    processed_feature_df = cleanup_results[1] 
    
    # Execute T Test 
    ttest_results = autobucket_t_test(processed_feature_df, processed_target_df)
     
    # Put the feature into the list for either pass or fail based on whether the p value is below the threshold 
    
    if ttest_results['p_Value'][0] < max_p_value: 
        features_passed_t_test.append(feature)
        pass_fail = 'PASS'
    else:
        features_failed_t_test.append(feature)
        pass_fail = 'FAIL'
    
    # Store feature t test results 
    # Add a row to the features t test results DataFrame
    feature_t_test_results.loc[len(feature_t_test_results)] = [feature, pass_fail, ttest_results['p_Value'][0], ttest_results['T_Statistic'][0], ttest_results['High_Target_Feature_Avg'][0], ttest_results['Low_Target_Feature_Avg'][0]]
    
print(f"Passed features: {features_passed_t_test}")
print(f"Failed features: {features_failed_t_test}")
```

```python
def execute_random_forest(target_series, feature_dataframe, estimator_count):

    #Check data type of target_series 
    if isinstance(target_series, pd.Series):
       
        # Check for non-null values
        non_null_values = target_series.isna().any()
        # Check for non-numeric values
        non_numeric_values = pd.to_numeric(target_series, errors='coerce').isnull().any()
        
        if non_null_values or non_numeric_values:
            proper_format = 0
            print(f"The target_series submitted to the execute_random_forest() contains non-null or non-numeric values")
            return
        else:
            proper_format = 1 
            
    else:
        proper_format = 0
        print(f"The target_series variable submitted to the execute_random_forest() function is not a Pandas series. It has data type: {type(target_series)}")
        return
    
    #Check data type of feature_dataframe
    if isinstance(feature_dataframe, pd.DataFrame):
        
        # Check for non-null values
        non_null_values = feature_dataframe.isna().any().any()
        # Check for non-numeric values
        non_numeric_values = feature_dataframe.applymap(lambda x: pd.to_numeric(x, errors='coerce')).isnull().any().any()
        
        if non_null_values or non_numeric_values:
            proper_format = 0
            print(f"The feature_dataframe submitted to the execute_random_forest() contains non-null or non-numeric values")
            return
        else:
            proper_format = 1 
                  
    else:
        proper_format = 0
        print(f"The feature_dataframe submitted to the execute_random_forest() function is not a Pandas dataframe. It has data type: {type(feature_dataframe)}")
        return
        
        
    # Check whether analysis should be classification or regression
    unique_values = target_series.unique()
    if len(unique_values) > 6:
        classification = 0
    else:
        classification = 1

    #Split data into test and training data sets
    X_train, X_test, y_train, y_test = train_test_split(feature_dataframe, target_series, test_size=0.3, random_state=42)

    # Use classification approach if appropriate 
    if classification == 1:
        # For classification
        clf = RandomForestClassifier(n_estimators=estimator_count, random_state=42) 
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        print("CLASSIFICATION Mean Squared Error:", mean_squared_error(y_test, y_pred))

    else:
        reg = RandomForestRegressor(n_estimators=estimator_count, random_state=42) 
        reg.fit(X_train, y_train)
        y_pred = reg.predict(X_test)
        print("REGRESSION Mean Squared Error:", mean_squared_error(y_test, y_pred))

    # Get the referense MSE    
    base_result_MSE = mean_squared_error(y_test, y_pred)


execute_random_forest(target_df, feature_df, 20)
```
